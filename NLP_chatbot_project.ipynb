```python
# ==========================================
# BIT4133: NLP Chatbot Project
# Business: Repair, Wholesale Parts, & Accessories
# ==========================================

# First, we need to import NLTK (Natural Language Toolkit)
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Download necessary NLTK data packages for tokenization and text cleaning
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Required for modern NLTK tokenizers in Colab

def clean_and_tokenize(user_input):
    """
    This function demonstrates Week 1 & 2 NLP Concepts:
    1. Tokenization: Breaking sentences into words.
    2. Lowercasing: Making everything lowercase for uniform matching.
    3. Stopword Removal: Removing low-meaning words (the, is, a) to focus on keywords.
    """
    # 1. Lowercase and Tokenize
    tokens = word_tokenize(user_input.lower())

    # 2. Get list of English stopwords and punctuation
    stop_words = set(stopwords.words('english'))
    punctuation = set(string.punctuation)

    # 3. Filter out stopwords and punctuation
    cleaned_tokens = [word for word in tokens if word not in stop_words and word not in punctuation]

    return cleaned_tokens

def get_chatbot_response(user_input):
    """
    Intent Matching System: Matches cleaned keywords to specific business services.
    """
    # Clean the input using our NLP function
    cleaned_words = clean_and_tokenize(user_input)

    # Define keywords for our business intents
    repair_keywords = ['repair', 'fix', 'screen', 'broken', 'battery', 'charging', 'mend']
    wholesale_keywords = ['wholesale', 'bulk', 'parts', 'spares', 'screens', 'displays', 'supply']
    accessory_keywords = ['accessory', 'accessories', 'case', 'charger', 'tempered', 'glass', 'pouch']
    greeting_keywords = ['hello', 'hi', 'hey', 'greetings', 'sup']

    # Check for greetings
    if any(word in cleaned_words for word in greeting_keywords):
        return ("👋 Welcome to Mobimend Spares! I am your AI assistant.\n"
                "How can I help you today? We offer:\n"
                "1. Professional Phone Repairs\n"
                "2. Wholesale Phone Parts & Spares\n"
                "3. High-Quality Phone Accessories")

    # Check for Repair intent
    elif any(word in cleaned_words for word in repair_keywords):
        return ("🔧 [REPAIR SERVICES]\n"
                "We fix broken screens, dead batteries, and charging ports for all major brands.\n"
                "Please bring your device to our service desk or reply with your phone model for a quote!")

    # Check for Wholesale intent
    elif any(word in cleaned_words for word in wholesale_keywords):
        return ("📦 [WHOLESALE PARTS & SPARES]\n"
                "We supply high-quality replacement screens, batteries, and charging flex cables in bulk.\n"
                "To request our latest wholesale price catalog, please leave your email address or business name.")

    # Check for Accessories intent
    elif any(word in cleaned_words for word in accessory_keywords):
        return ("🎧 [ACCESSORIES]\n"
                "Protect your phone! We stock premium tempered glass protectors, durable phone cases, fast chargers, and wireless earphones.\n"
                "Discounts available on bundle purchases!")

    # Fallback response if keywords don't match
    else:
        return ("🤔 I'm not quite sure I understood that. You can ask me about 'phone repairs', "
                "'wholesale spare parts', or 'phone accessories'!")

# ==========================================
# CHATBOT SIMULATION INTERFACE
# ==========================================
print("🤖 Chatbot initialized! Type 'exit' to stop the chat.\n")

# Simple loop to keep the chat running in Colab
while True:
    user_msg = input("You: ")
    if user_msg.lower() == 'exit':
        print("Chatbot: Goodbye! Thank you for contacting Mobimend Spares.")
        break

    # Run the processing and print response
    response = get_chatbot_response(user_msg)
    print(f"Chatbot: {response}\n")
```
