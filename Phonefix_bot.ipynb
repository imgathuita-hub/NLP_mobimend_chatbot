```python
from nltk.corpus import stopwords
import string  # For punctuation removal

# Load English stopwords (NLTK has 179 of them)
stop_words = set(stopwords.words('english'))

print(f"Total English stopwords in NLTK: {len(stop_words)}")
print(f"Sample stopwords: {sorted(list(stop_words))[:20]}")

# Function to clean customer messages
def preprocess_message(message):
    """
    This function:
    1. Lowercases all text (so 'Samsung' = 'samsung')
    2. Tokenizes the message into words
    3. Removes stopwords
    4. Removes punctuation
    Returns only meaningful keyword tokens.
    """
    # Step 1: Lowercase everything
    message_lower = message.lower()

    # Step 2: Tokenize
    tokens = word_tokenize(message_lower)

    # Step 3: Remove stopwords AND punctuation
    keywords = [
        word for word in tokens
        if word not in stop_words          # not a stopword
        and word not in string.punctuation  # not punctuation
    ]

    return keywords

# Test with real customer messages
test_messages = [
    "How much does a Samsung screen repair cost?",
    "Do you sell iPhone 15 batteries wholesale?",
    "My phone fell in water. Can you fix water damage?",
    "What accessories do you have for Tecno phones?",
]

print("\n" + "=" * 60)
print("📌 STOPWORD REMOVAL OUTPUT")
print("=" * 60)

for msg in test_messages:
    keywords = preprocess_message(msg)
    print(f"\n📥 Original : '{msg}'")
    print(f"🔑 Keywords : {keywords}")
```

```python
# Import everything we need
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.util import ngrams
from nltk import pos_tag
from collections import Counter
import string

# Download required NLTK data (safe to re-run)
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('punkt_tab', quiet=True)

stop_words = set(stopwords.words('english'))

print("✅ All imports ready!")
```

```python
# === STEP 1: Generate N-grams from Customer Messages ===

def get_ngrams(text, n):
    """
    Extract n-grams from a text string.
    Steps:
    1. Lowercase
    2. Tokenize
    3. Remove punctuation
    4. Generate n-grams using nltk.util.ngrams
    """
    tokens = word_tokenize(text.lower())
    tokens = [t for t in tokens if t not in string.punctuation]
    return list(ngrams(tokens, n))

# Real customer messages for our phone repair chatbot
sample_messages = [
    "How much does a Samsung screen repair cost?",
    "I need iPhone 14 battery replacement",
    "Do you do water damage repair for Samsung?",
    "I want to buy phone cases in bulk wholesale",
    "My screen is cracked and touch is not working",
]

print("=" * 60)
print("📌 BIGRAM OUTPUT (N=2)")
print("=" * 60)
for msg in sample_messages:
    bigrams = get_ngrams(msg, 2)
    print(f"\n💬 '{msg}'")
    print(f"   Bigrams: {bigrams}")

print("\n" + "=" * 60)
print("📌 TRIGRAM OUTPUT (N=3)")
print("=" * 60)
for msg in sample_messages:
    trigrams = get_ngrams(msg, 3)
    print(f"\n💬 '{msg}'")
    print(f"   Trigrams: {trigrams}")
```

```python
# === STEP 2: Most Common Bigrams in Customer Queries ===
# This is how real chatbots learn what customers ask most!

# Simulated customer query history (what customers asked last month)
customer_query_history = [
    "screen repair samsung",
    "iphone screen replacement",
    "battery replacement cost",
    "samsung screen crack repair",
    "wholesale phone parts order",
    "water damage repair iphone",
    "screen protector price",
    "phone case tecno infinix",
    "battery replacement iphone",
    "screen repair price samsung",
    "iphone battery cost",
    "wholesale battery order",
    "screen replacement cost",
    "water damage samsung fix",
    "charger cable usb type",
]

# Collect all bigrams from all queries
all_bigrams = []
for query in customer_query_history:
    all_bigrams.extend(get_ngrams(query, 2))

# Count most frequent
bigram_freq = Counter(all_bigrams)
top_bigrams = bigram_freq.most_common(10)

print("=" * 55)
print("📊 TOP 10 MOST COMMON CUSTOMER QUERY BIGRAMS")
print("   (What customers ask about most often)")
print("=" * 55)
for i, (bigram, count) in enumerate(top_bigrams, 1):
    bar = "█" * count
    print(f"  {i:2}. {' '.join(bigram):<25} {bar} ({count}x)")
```

```python
# === POS TAGGING ===

def pos_tag_message(message):
    """Tag each word in a message with its part of speech."""
    tokens = word_tokenize(message)
    tagged = pos_tag(tokens)  # Returns list of (word, POS_tag) tuples
    return tagged

# Full POS tag descriptions for display
TAG_DESCRIPTIONS = {
    'NN': 'Noun', 'NNS': 'Noun(plural)', 'NNP': 'Proper Noun',
    'VB': 'Verb', 'VBZ': 'Verb(3rdperson)', 'VBD': 'Verb(past)',
    'JJ': 'Adjective', 'JJR': 'Adj(comparative)',
    'RB': 'Adverb', 'WRB': 'Wh-adverb',
    'CD': 'Number', 'DT': 'Determiner',
    'IN': 'Preposition', 'PRP': 'Pronoun',
    'WP': 'Wh-pronoun', 'MD': 'Modal(can/will)',
    'CC': 'Conjunction', '.': 'Punctuation', '?': 'Punctuation'
}

test_sentences = [
    "How much does a Samsung screen repair cost?",
    "I need 20 iPhone batteries for wholesale order",
    "My cracked screen is not working properly",
    "Can you fix water damaged Tecno phone?",
]

print("=" * 60)
print("📌 POS TAGGING OUTPUT")
print("=" * 60)

for sentence in test_sentences:
    tagged = pos_tag_message(sentence)
    print(f"\n💬 '{sentence}'")
    print(f"   {'WORD':<18} {'TAG':<8} MEANING")
    print(f"   {'----':<18} {'---':<8} -------")
    for word, tag in tagged:
        desc = TAG_DESCRIPTIONS.get(tag, tag)
        print(f"   {word:<18} {tag:<8} {desc}")
```

```python
# === USE POS TAGS TO EXTRACT STRUCTURED INFORMATION ===
# This is how chatbots extract: Brand + Part + Problem from a single sentence

KNOWN_BRANDS = {'samsung', 'iphone', 'apple', 'tecno', 'infinix', 'nokia',
                'huawei', 'xiaomi', 'redmi', 'oppo', 'vivo', 'realme'}

KNOWN_PARTS = {'screen', 'battery', 'charger', 'port', 'camera',
               'speaker', 'microphone', 'button', 'display', 'lcd'}

PROBLEM_WORDS = {'cracked', 'broken', 'damaged', 'dead', 'slow',
                 'overheating', 'wet', 'swollen', 'faulty', 'not'}

def extract_entities(message):
    """
    Extract brand, part, and problem from a customer message
    using both keywords AND POS tags.
    """
    tokens = word_tokenize(message.lower())
    tagged = pos_tag(tokens)

    found = {'brand': [], 'part': [], 'problem': [], 'quantity': []}

    for word, tag in tagged:
        clean = word.lower()
        if clean in KNOWN_BRANDS:
            found['brand'].append(word)
        if clean in KNOWN_PARTS:
            found['part'].append(word)
        if clean in PROBLEM_WORDS:
            found['problem'].append(word)
        if tag == 'CD':  # Cardinal number
            found['quantity'].append(word)

    return found

test_messages = [
    "How much does a Samsung screen repair cost?",
    "I need 20 iPhone batteries for wholesale",
    "My Tecno cracked screen is not working",
    "Can you fix my Samsung broken charger port?",
    "I want to order 50 Infinix batteries",
]

print("=" * 60)
print("📌 ENTITY EXTRACTION FROM POS TAGS")
print("   (Brand + Part + Problem + Quantity)")
print("=" * 60)

for msg in test_messages:
    entities = extract_entities(msg)
    print(f"\n💬 '{msg}'")
    print(f"   🏷️  Brand   : {entities['brand'] or 'Not specified'}")
    print(f"   🔧 Part    : {entities['part'] or 'Not specified'}")
    print(f"   ⚠️  Problem : {entities['problem'] or 'Not specified'}")
    print(f"   🔢 Quantity: {entities['quantity'] or 'Not specified'}")
```

```python
# === PhoneFixBot v2.0 — N-gram + POS aware ===

# Re-use our functions from earlier in this notebook

# Bigram-based intents (phrases that signal specific needs)
BIGRAM_INTENTS = {
    ('screen', 'repair')    : 'screen_repair',
    ('screen', 'replacement'): 'screen_repair',
    ('battery', 'replacement'): 'battery_repair',
    ('battery', 'replacement'): 'battery_repair',
    ('water', 'damage')     : 'water_damage',
    ('wholesale', 'order')  : 'wholesale',
    ('bulk', 'order')       : 'wholesale',
    ('phone', 'case')       : 'accessories',
    ('screen', 'protector') : 'accessories',
}

RESPONSES = {
    'screen_repair': "📱 Screen repair prices start from KSh 1,500. Send your phone model for exact pricing!",
    'battery_repair': "🔋 Battery replacement from KSh 1,200. We stock batteries for all major brands!",
    'water_damage': "💧 Bring your phone in immediately! Do NOT charge it. Free diagnosis.",
    'wholesale': "📦 We offer wholesale pricing for 5+ units. WhatsApp us for the price list!",
    'accessories': "🛍️ We stock cases, tempered glass, chargers and cables. Shop online at our website!",
}

def chatbot_v2(user_message):
    """PhoneFixBot v2.0 using bigrams + POS-based entity extraction."""
    # 1. Extract bigrams from message
    msg_bigrams = set(get_ngrams(user_message, 2))

    # 2. Check bigram intents first (more accurate than single keywords)
    for bigram, intent in BIGRAM_INTENTS.items():
        if bigram in msg_bigrams:
            response = RESPONSES[intent]
            # 3. Personalise with POS-extracted brand and quantity
            entities = extract_entities(user_message)
            extra = ""
            if entities['brand']:
                extra += f" (for {', '.join(entities['brand'])})"
            if entities['quantity']:
                extra += f" [Quantity noted: {', '.join(entities['quantity'])} units]"
            return response + extra

    # 4. Fallback: single keyword check
    tokens = set(word_tokenize(user_message.lower()))
    if tokens & {'screen', 'cracked', 'display', 'lcd'}:
        return RESPONSES['screen_repair']
    if tokens & {'battery', 'charge', 'charging'}:
        return RESPONSES['battery_repair']
    if tokens & {'water', 'wet', 'liquid'}:
        return RESPONSES['water_damage']

    return "🤖 I didn't catch that. Try: screen repair, battery, water damage, wholesale, or accessories."


# Test upgraded chatbot
test_conversations = [
    "How much does Samsung screen repair cost?",
    "I need iPhone battery replacement",
    "I want to do a wholesale order for 30 Tecno batteries",
    "Do you sell screen protectors?",
    "My phone has water damage",
]

print("=" * 60)
print("🤖 PhoneFixBot v2.0 — N-gram + POS Upgrade")
print("=" * 60)
for user_msg in test_conversations:
    print(f"\n👤 Customer: {user_msg}")
    print(f"🤖 Bot: {chatbot_v2(user_msg)}")
```

```python
# Install and import all required libraries
!pip install nltk spacy -q
!python -m spacy download en_core_web_sm -q

import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
from nltk.tree import Tree
import re

nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('maxent_ne_chunker', quiet=True)
nltk.download('words', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('maxent_ne_chunker_tab', quiet=True) # Added missing download

# Load spaCy model (more powerful NER)
nlp = spacy.load("en_core_web_sm")

print("✅ All libraries loaded!")
```

```python
# === CUSTOM SEQUENCE LABELER FOR PHONEFIX DOMAIN ===
# We define our own entity categories relevant to phone repair

# Our custom entity vocabulary
ENTITY_VOCAB = {
    # Format: 'word_lowercase': 'ENTITY_TYPE'
    # Phone brands
    'iphone': 'BRAND', 'samsung': 'BRAND', 'tecno': 'BRAND',
    'infinix': 'BRAND', 'nokia': 'BRAND', 'huawei': 'BRAND',
    'xiaomi': 'BRAND', 'redmi': 'BRAND', 'oppo': 'BRAND',

    # Phone parts
    'screen': 'PART', 'display': 'PART', 'lcd': 'PART',
    'battery': 'PART', 'charger': 'PART', 'port': 'PART',
    'camera': 'PART', 'speaker': 'PART', 'button': 'PART',
    'cable': 'PART', 'case': 'PART', 'cover': 'PART',

    # Services
    'repair': 'SERVICE', 'fix': 'SERVICE', 'replace': 'SERVICE',
    'replacement': 'SERVICE', 'clean': 'SERVICE',

    # Problems
    'cracked': 'PROBLEM', 'broken': 'PROBLEM', 'damaged': 'PROBLEM',
    'water': 'PROBLEM', 'dead': 'PROBLEM', 'slow': 'PROBLEM',
    'overheating': 'PROBLEM', 'wet': 'PROBLEM',

    # Business types
    'wholesale': 'ORDER_TYPE', 'bulk': 'ORDER_TYPE', 'retail': 'ORDER_TYPE',

    # Locations (Kenyan cities for our context)
    'nairobi': 'LOCATION', 'mombasa': 'LOCATION', 'kisumu': 'LOCATION',
    'nakuru': 'LOCATION', 'eldoret': 'LOCATION',
}

def bio_sequence_label(sentence):
    """
    Apply BIO sequence labeling to a sentence.
    Returns a list of (token, BIO_label) tuples.
    """
    tokens = word_tokenize(sentence)
    labels = []
    prev_entity_type = None

    for token in tokens:
        token_lower = token.lower()

        # Check if token is a number (quantity)
        if token.isdigit():
            entity_type = 'QUANTITY'
        else:
            entity_type = ENTITY_VOCAB.get(token_lower, None)

        if entity_type is None:
            labels.append((token, 'O'))
            prev_entity_type = None
        elif entity_type == prev_entity_type:
            # Continuation of same entity type → I- tag
            labels.append((token, f'I-{entity_type}'))
        else:
            # Beginning of new entity → B- tag
            labels.append((token, f'B-{entity_type}'))
            prev_entity_type = entity_type

    return labels


# Test sentences for our phone repair chatbot
test_sentences = [
    "How much does Samsung screen repair cost?",
    "I need 20 iPhone batteries for wholesale order",
    "My cracked Tecno display needs replacement",
    "Can you fix water damaged Samsung phone in Nairobi?",
    "I want to buy 50 Infinix charger cables bulk",
]

print("=" * 65)
print("📌 BIO SEQUENCE LABELING OUTPUT")
print("=" * 65)

for sentence in test_sentences:
    labeled = bio_sequence_label(sentence)
    print(f"\n💬 '{sentence}'")
    print(f"   {'TOKEN':<18} LABEL")
    print(f"   {'-----':<18} -----")
    for token, label in labeled:
        marker = "  ◀" if label != 'O' else ""
        print(f"   {token:<18} {label}{marker}")
```

```python
# === SPACY NER ===

def ner_with_spacy(sentence):
    """Run spaCy NER on a sentence and display results."""
    doc = nlp(sentence)
    return [(ent.text, ent.label_, spacy.explain(ent.label_)) for ent in doc.ents]

test_sentences_ner = [
    "I need 20 batteries shipped to Nairobi by Friday for KSh 2000 each",
    "Can Samsung repair an iPhone 14 Pro screen?",
    "John from Mombasa ordered 100 Xiaomi parts last Tuesday",
    "The repair shop in Westlands quoted KSh 3500 for a screen fix",
]

print("=" * 65)
print("📌 SPACY NAMED ENTITY RECOGNITION OUTPUT")
print("=" * 65)

for sentence in test_sentences_ner:
    entities = ner_with_spacy(sentence)
    print(f"\n💬 '{sentence}'")
    if entities:
        print(f"   {'ENTITY TEXT':<25} {'LABEL':<12} EXPLANATION")
        print(f"   {'-----------':<25} {'-----':<12} -----------")
        for text, label, explanation in entities:
            print(f"   {text:<25} {label:<12} {explanation}")
    else:
        print("   (No standard named entities detected)")
```

```python
# === NLTK NE_CHUNK (Alternative NER method) ===
# NLTK uses a tree structure to group named entities

def nltk_ner(sentence):
    """Use NLTK's built-in NE chunker for NER."""
    tokens = word_tokenize(sentence)
    tagged = pos_tag(tokens)
    chunked = ne_chunk(tagged)

    named_entities = []
    for subtree in chunked:
        if isinstance(subtree, Tree):  # It's a named entity (multi-word)
            entity = " ".join([token for token, pos in subtree.leaves()])
            entity_type = subtree.label()
            named_entities.append((entity, entity_type))

    return named_entities

print("=" * 60)
print("📌 NLTK NE_CHUNK OUTPUT")
print("=" * 60)

nltk_test_sentences = [
    "Samsung phones are repaired at our Nairobi branch",
    "John ordered 50 Apple parts from our wholesale store",
    "Our shop is located in Westlands, Nairobi, Kenya",
]

for sentence in nltk_test_sentences:
    entities = nltk_ner(sentence)
    print(f"\n💬 '{sentence}'")
    if entities:
        for entity, label in entities:
            print(f"   ▶ '{entity}' → {label}")
    else:
        print("   (No named entities found by NLTK)")
```

```python
# === PhoneFixBot v3.0 — Full NLP Pipeline ===

from nltk.corpus import stopwords
from nltk.util import ngrams as nltk_ngrams
import string

stop_words = set(stopwords.words('english'))

def full_nlp_pipeline(message):
    """
    Complete NLP pipeline for PhoneFixBot v3.0

    Stage 1 (Week 1): Tokenize + remove stopwords
    Stage 2 (Week 2): Extract bigrams + POS tags
    Stage 3 (Week 3): BIO sequence label + NER

    Returns a structured intent dict.
    """
    result = {
        'original': message,
        'keywords': [],
        'bigrams': [],
        'entities': {'brand': [], 'part': [], 'problem': [], 'quantity': [], 'location': []},
        'intent': 'unknown',
    }

    # ── STAGE 1: Tokenize + stopwords ────────────────────────
    tokens = word_tokenize(message.lower())
    result['keywords'] = [
        t for t in tokens
        if t not in stop_words and t not in string.punctuation
    ]

    # ── STAGE 2: N-grams + POS ───────────────────────────────
    result['bigrams'] = list(nltk_ngrams(result['keywords'], 2))
    pos_tagged = pos_tag(word_tokenize(message))

    # ── STAGE 3: BIO labeling + entity extraction ────────────
    labeled = bio_sequence_label(message)
    for token, label in labeled:
        if 'BRAND' in label:    result['entities']['brand'].append(token)
        if 'PART' in label:     result['entities']['part'].append(token)
        if 'PROBLEM' in label:  result['entities']['problem'].append(token)
        if 'QUANTITY' in label: result['entities']['quantity'].append(token)
        if 'LOCATION' in label: result['entities']['location'].append(token)

    # ── INTENT CLASSIFICATION ────────────────────────────────
    kw_set = set(result['keywords'])
    bigram_set = set(result['bigrams'])

    if ('screen', 'repair') in bigram_set or ('screen', 'replacement') in bigram_set:
        result['intent'] = 'screen_repair'
    elif ('battery', 'replacement') in bigram_set or 'battery' in kw_set:
        result['intent'] = 'battery_repair'
    elif ('water', 'damage') in bigram_set or 'water' in kw_set:
        result['intent'] = 'water_damage'
    elif 'wholesale' in kw_set or 'bulk' in kw_set:
        result['intent'] = 'wholesale'
    elif kw_set & {'case', 'cover', 'accessory', 'accessories', 'charger'}:
        result['intent'] = 'accessories'
    elif kw_set & {'price', 'cost', 'much', 'fee'}:
        result['intent'] = 'pricing'
    elif kw_set & {'hello', 'hi', 'hey', 'help'}:
        result['intent'] = 'greeting'

    return result


def generate_response_v3(analysis):
    """Generate a personalised response using extracted entities."""
    intent = analysis['intent']
    brand = analysis['entities']['brand']
    part = analysis['entities']['part']
    qty = analysis['entities']['quantity']
    loc = analysis['entities']['location']

    brand_str = f" for {brand[0]}" if brand else ""
    qty_str = f" ({qty[0]} units)" if qty else ""

    base_responses = {
        'screen_repair': f"📱 {brand[0] + ' s' if brand else 'S'}creen repair starts from KSh 1,500{brand_str}. Walk in or book an appointment!",
        'battery_repair': f"🔋 Battery replacement from KSh 1,200{brand_str}. We stock all major brands!",
        'water_damage': f"💧 Bring the phone in immediately — do NOT charge it! Free diagnosis.",
        'wholesale': f"📦 Wholesale pricing available{qty_str}. WhatsApp +254 700 000 000 for the full price list!",
        'accessories': f"🛍️ We stock cases, screen protectors, chargers, and cables{brand_str}!",
        'pricing': f"💰 Prices range from KSh 1,200 – KSh 6,500 depending on the model{brand_str}. Send your model for an exact quote!",
        'greeting': "👋 Welcome to PhoneFixBot! I can help with screen repair, battery, water damage, wholesale, or accessories. What do you need?",
        'unknown': "🤖 I didn't quite catch that. Try asking about: screen repair, battery, water damage, wholesale, or accessories.",
    }
    return base_responses.get(intent, base_responses['unknown'])


# ── FINAL CHATBOT DEMO ────────────────────────────────────────
final_test = [
    "Hello, I need help",
    "How much does Samsung screen repair cost?",
    "I need 50 iPhone batteries for wholesale",
    "My cracked Tecno display needs replacement",
    "My phone has water damage, it fell in the sink",
    "What phone cases do you have for Infinix?",
]

print("=" * 65)
print("🤖 PhoneFixBot v3.0 — Full NLP Pipeline Demo")
print("   (Tokenization + N-grams + POS + NER)")
print("=" * 65)

for msg in final_test:
    analysis = full_nlp_pipeline(msg)
    response = generate_response_v3(analysis)

    print(f"\n👤 Customer : {msg}")
    print(f"   Intent   : {analysis['intent']}")
    print(f"   Entities : Brand={analysis['entities']['brand']}, "
          f"Part={analysis['entities']['part']}, "
          f"Qty={analysis['entities']['quantity']}")
    print(f"🤖 Bot      : {response}")
```
